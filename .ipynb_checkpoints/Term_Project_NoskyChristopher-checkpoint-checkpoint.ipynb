```python
clean_hr = pd.DataFrame(hr_data, columns = ['relevent_experience', 'education_level', 'major_discipline',
                                              'experience', 'company_size', 'company_type', 'last_new_job'])
```

```python
clean_hr.dropna(how='any', inplace=True)
```

```python
if clean_hr.loc[clean_hr['experience'].str.startswith('<')] is not None:
    clean_hr.replace(to_replace='<1', value='1', inplace = True)
```

```python
if clean_hr.loc[clean_hr['experience'].str.startswith('>')] is not None:
    clean_hr.replace(to_replace='>20', value='20', inplace = True)
```

```python
if clean_hr.loc[clean_hr['last_new_job'].str.startswith('<')] is not None:
    clean_hr.replace(to_replace='>4', value='4', inplace=True)
```

```python
if clean_hr.loc[clean_hr['last_new_job'].str.startswith('never')] is not None:
    clean_hr.replace(to_replace='never', value='0', inplace=True)
```

## CDF

```python
def evalCdf(t,x):
    count = 0.0
    for value in t:
        if value <=x:
            count += 1
            
    prob = count/len(t)
    return prob
```

```python
def cdf_func(clean_hr):
    experience = clean_hr.experience
    cdf = thinkstats2.Cdf(experience, label='experience')

    sample = cdf.Sample(1000)
    sample_cdf = thinkstats2.Cdf(sample, label='sample')

    thinkplot.PrePlot(2)
    thinkplot.Cdfs([cdf, sample_cdf])
    thinkplot.Save(root='cumulative_sample',
                       xlabel='Experience (Years)',
                       ylabel='CDF')
```

```python

cdf_func
```

```python
cdf=thinkstats2.Cdf(clean_hr.experience, label='Experience')
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='Experience', ylabel='CDF')
```

```python
def MakeNormalPlot(clean_hr):
    exp_mean = clean_hr.experience.mean()
    exp_std = clean_hr.experience.std()


    xs = [-2, 2]
    fxs, fys = thinkstats2.FitLine(xs, inter=exp_mean, slope=exp_std)
    thinkplot.Plot(fxs, fys, color='gray', label='model')

    xs, ys = thinkstats2.NormalProbability(clean_hr['experience'])
    thinkplot.Plot(xs, ys, label='experience')

```

```python
MakeNormalPlot(clean_hr)
```

```python
from statsmodels.stats import weightstats as stests

ztest ,pval = stests.ztest(clean_hr['experience'], x2=None, value=156)

print(float(pval))

if pval<0.05:
    
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
ztest ,pval = stests.ztest(clean_hr['last_new_job'], x2=None, value=156)

print(float(pval))

if pval<0.05:
    
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
